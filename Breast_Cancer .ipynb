#**EDA**

```python
diagnosis_counts = df['diagnosis'].value_counts()

print(diagnosis_counts)

diagnosis_counts.plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title('Distribution of Diagnosis (M vs B)')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

**boxplots visualize the relationship between the features 'radius_mean' and 'perimeter_mean' with the diagnosis type **



```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='diagnosis', y='radius_mean', data=df)
plt.title('Boxplot of Radius Mean by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Radius Mean')

plt.subplot(1, 2, 2)
sns.boxplot(x='diagnosis', y='perimeter_mean', data=df)
plt.title('Boxplot of Perimeter Mean by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Perimeter Mean')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.boxplot(x='diagnosis', y='smoothness_mean', data=df)
plt.title('Smoothness Mean by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Smoothness Mean')


plt.subplot(1, 2, 2)
sns.boxplot(x='diagnosis', y='concavity_mean', data=df)
plt.title('Concavity Mean by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Concavity Mean')

plt.tight_layout()
plt.show()
```

```python
# Encode the diagnosis (M -> 1, B -> 0) for easier plotting
df['diagnosis'] = df['diagnosis'].apply(lambda x: 'Malignant' if x == 'M' else 'Benign')

# Create a new DataFrame with only relevant columns
radius_data = df[['diagnosis', 'radius_mean', 'radius_worst']]

# Melt the DataFrame to have 'radius_mean' and 'radius_worst' in a single column for plotting
radius_melted = radius_data.melt(id_vars='diagnosis', var_name='Radius Type', value_name='Value')

# Plot the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Radius Type', y='Value', hue='diagnosis', data=radius_melted, palette='deep')
plt.title('Comparison of Radius Mean and Radius Worst by Diagnosis')
plt.ylabel('Radius Value')
plt.show()
```
